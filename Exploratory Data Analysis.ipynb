```python
df_o = df_envanter[df_envanter["Değişen Parça Replacibility"] == "O"]
next_map = dict(zip(df_o["Malzeme Kodu"].astype(str), df_o["Değişen Malzeme Kodu"].astype(str)))
prev_map = {v: k for k, v in next_map.items()}

# Geriye yürüyerek zincirin kökünü bul
def find_root(code):
    while code in prev_map:
        code = prev_map[code]
    return code

# İleriye yürüyerek zinciri oluştur
def walk_chain(root):
    zincir = []
    while root not in zincir:
        zincir.append(root)
        if root not in next_map:
            break
        root = next_map[root]
    return zincir

# Tüm zincirleri oluştur ve zincirdeki her elemana yaz
zincir_map = {}
tum_kodlar = set(next_map.keys()) | set(next_map.values())

for kod in tum_kodlar:
    kok = find_root(kod)
    zincir = walk_chain(kok)
    for z in zincir:
        if z not in zincir_map or len(zincir) > len(zincir_map[z]):
            zincir_map[z] = zincir

# Sonuçları yaz
df_envanter["Malzeme Kodu"] = df_envanter["Malzeme Kodu"].astype(str)
df_envanter["Degisen Malzeme List"] = df_envanter["Malzeme Kodu"].map(zincir_map)
```

### Standart maliyet için değişen parça farkları

```python
import pandas as pd
import numpy as np

# 1) Yardımcılar
def normalize_code(x):
    if x is None:
        return None
    s = str(x).strip()
    if s.endswith(".0"):
        s = s[:-2]
    return s

def parse_code_list(s):
    # NaN/None
    if s is None:
        return []
    if isinstance(s, float) and pd.isna(s):
        return []
    # Liste/ndarray gelirse birleştir
    if isinstance(s, (list, np.ndarray)):
        s = ",".join(map(str, s))
    # String olarak temizle
    s = str(s).strip()
    if s == "" or s.lower() in ["nan", "none"]:
        return []
    s = s.strip("[]")                    # "[a, b, c]" durumunu düzelt
    parts = [normalize_code(p) for p in s.split(",")]
    parts = [p for p in parts if p and p.strip() != ""]
    return parts

# 2) df_envanter_2'den Kod -> Standart maliyet map'i
tmp = df_envanter2.copy()
tmp["Malzeme Kodu"] = tmp["Malzeme Kodu"].apply(normalize_code).astype(str)
maliyet_map = (
    tmp.drop_duplicates(subset=["Malzeme Kodu"])
       .set_index("Malzeme Kodu")["Standart maliyet"]
)

# 3) Sadece 'Degisen Malzeme List' dolu satırlar + kod listesine çevir
work = df_envanter.copy()
work["row_id"] = work.index
work["Kodlar"] = work["Degisen Malzeme List"].apply(parse_code_list)
work = work[work["Kodlar"].apply(lambda lst: len(lst) > 0)].copy()

# 4) Explode edip maliyeti eşle
expl = work.explode("Kodlar").rename(columns={"Kodlar":"Kod"})
expl["Kod"] = expl["Kod"].astype(str)
expl["Standart_maliyet"] = expl["Kod"].map(maliyet_map)

# 5) Satır bazında min, max, fark (en az 2 geçerli maliyet varsa fark hesapla)
g = expl.groupby("row_id")["Standart_maliyet"]
summary = pd.DataFrame({
    "Min_Maliyet": g.min(),
    "Maks_Maliyet": g.max(),
    "N_gecerli": g.count()
})
summary["Maliyet_Fark"] = np.where(summary["N_gecerli"] >= 2,
                                   summary["Maks_Maliyet"] - summary["Min_Maliyet"],
                                   np.nan)

# (Opsiyonel) Hangi kodlar ve hangi maliyetler kullanıldı?
lists = expl.groupby("row_id").agg(
    Kodlar_list=("Kod", list),
    Maliyetler=("Standart_maliyet", lambda s: [float(x) for x in s.dropna()])
)

# 6) Sonuç tablosu: sadece dolu satırlar + farklar
df_sonuc = (work.set_index("row_id")
                 .join(summary)
                 .join(lists)
                 .reset_index(drop=True))

df_sonuc
```
